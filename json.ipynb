```python
import os

path = '/content/NTCIR14-QALab-PoliInfo-FormalRunDataset-master/Summarization/training'
```

```python
import pandas as pd
import json
df = []
for f in os.listdir(path):
  df.append(pd.read_json(path + '/' + f))
```

```python
DF = pd.concat(df)
```

```python
DF.loc[:,['Summary','Source']].to_csv('output.txt')
```

```python
DF['Summary']
```

```python
print(DF['Summary'][0])
```

```python
import numpy as np


# 列Sourceをシャッフルし、同じ位置に同じ要素が来ないようにする
shuffled_Source = DF['Source'].sample(frac=1).reset_index(drop=True)
while any(DF['Source'] == shuffled_Source):
    shuffled_Source = DF['Source'].sample(frac=1).reset_index(drop=True)

# シャッフル後のDataFrameを作成
print(shuffled_Source)
```

```python
print(DF['Source'])
```

```python
True_df = DF.loc[:,['Summary','Source']]
True_df['TF'] = 1
True_df
```

```python
False_df = pd.DataFrame({'Summary': DF['Summary'], 'Source': shuffled_Source})
False_df['TF'] = 0
False_df
```

```python
TF_df = pd.concat([True_df, False_df], axis=0, ignore_index=True)
TF_df
```

```python
TF_df.to_csv('dataset.txt')
```
